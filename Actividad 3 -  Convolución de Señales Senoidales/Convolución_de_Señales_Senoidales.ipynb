La actividad realizada consiste en:
- Se generan dos señales de entrada: una combinación de dos senoidales con frecuencias ajustables y una tercera señal senoidal independiente.
- Luego, se calcula la convolución entre ambas señales y se representa gráficamente junto con sus componentes individuales.
- Adiconalmente, se muestra una versión recortada de la convolución para resaltar una sección específica del resultado. Esta herramienta permite analizar el efecto de la convolución en el dominio del tiempo y su impacto en la superposición de señales.

```python
# Importamos librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

# Función interactiva que permite modificar las frecuencias de las señales
@interact(f1=(0, 50, 1), f2=(0, 50, 1), f3=(0, 50, 1))
def sin(f1=10, f2=6, f3=4):
    # Definir el vector de tiempo
    t = np.linspace(0, 1, 1000)

    # Definir las señales de entrada
    s1 = np.sin(2 * np.pi * t * f1) + np.sin(2 * np.pi * t * f2)  # Señal compuesta por dos senoidales
    s2 = np.sin(2 * np.pi * f3 * t)  # Señal senoidal simple

    # Convolución de las dos señales
    conv = np.convolve(s1, s2)
    t_2 = np.linspace(0, 1, len(conv))  # Ajuste del tiempo para la convolución

    # Crear una versión modificada de la convolución con valores limitados en el tiempo
    conv_2 = np.zeros_like(conv)
    conv_2[t_2 <= 0.2] = conv[t_2 <= 0.2]

    # Crear la figura y subgráficos
    fig = plt.figure(figsize=(10, 6))
    ax1 = plt.subplot2grid((2, 2), (0, 0), colspan=1)  # Señal 1
    ax2 = plt.subplot2grid((2, 2), (0, 1), colspan=1)  # Señal 2
    ax3 = plt.subplot2grid((2, 2), (1, 0), colspan=1)  # Convolución
    ax4 = plt.subplot2grid((2, 2), (1, 1), colspan=1)  # Convolución recortada

    # Graficar la señal 1
    ax1.plot(t, s1, color="r", label="$\sin(2\pi f_1 t) + \sin(2\pi f_2 t)$")
    ax1.set_title("Señal 1")
    ax1.set_xlabel("Tiempo (s)")
    ax1.set_ylabel("Amplitud")
    ax1.legend(loc="lower left")
    ax1.grid(True)

    # Graficar la señal 2
    ax2.plot(t, s2, color="b", label="$\sin(2\pi f_3 t)$")
    ax2.set_title('Señal 2')
    ax2.set_xlabel("Tiempo (s)")
    ax2.set_ylabel("Amplitud")
    ax2.legend()
    ax2.grid(True)

    # Graficar la convolución
    ax3.plot(t_2, conv, color="g")
    ax3.set_title('Convolución')
    ax3.set_xlabel("Tiempo (s)")
    ax3.set_ylabel("Amplitud")
    ax3.grid(True)

    # Graficar la convolución recortada
    ax4.plot(t_2, conv_2, color="black")
    ax4.set_title('Convolución recortada en el tiempo')
    ax4.set_xlim(0, 1)
    ax4.set_xlabel("Tiempo (s)")
    ax4.set_ylabel("Amplitud")

    # Ajustar diseño y mostrar
    plt.tight_layout()
    plt.show()
```
