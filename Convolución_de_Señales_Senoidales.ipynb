La actividad realizada consiste en:
- Se generan dos seales de entrada: una combinacin de dos senoidales con frecuencias ajustables y una tercera seal senoidal independiente.
- Luego, se calcula la convolucin entre ambas seales y se representa grficamente junto con sus componentes individuales.
- Adiconalmente, se muestra una versin recortada de la convolucin para resaltar una seccin especfica del resultado. Esta herramienta permite analizar el efecto de la convolucin en el dominio del tiempo y su impacto en la superposicin de seales.

```python
# Importamos librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

# Funcin interactiva que permite modificar las frecuencias de las seales
@interact(f1=(0, 50, 1), f2=(0, 50, 1), f3=(0, 50, 1))
def sin(f1=10, f2=6, f3=4):
    # Definir el vector de tiempo
    t = np.linspace(0, 1, 1000)

    # Definir las seales de entrada
    s1 = np.sin(2 * np.pi * t * f1) + np.sin(2 * np.pi * t * f2)  # Seal compuesta por dos senoidales
    s2 = np.sin(2 * np.pi * f3 * t)  # Seal senoidal simple

    # Convolucin de las dos seales
    conv = np.convolve(s1, s2)
    t_2 = np.linspace(0, 1, len(conv))  # Ajuste del tiempo para la convolucin

    # Crear una versin modificada de la convolucin con valores limitados en el tiempo
    conv_2 = np.zeros_like(conv)
    conv_2[t_2 <= 0.2] = conv[t_2 <= 0.2]

    # Crear la figura y subgrficos
    fig = plt.figure(figsize=(10, 6))
    ax1 = plt.subplot2grid((2, 2), (0, 0), colspan=1)  # Seal 1
    ax2 = plt.subplot2grid((2, 2), (0, 1), colspan=1)  # Seal 2
    ax3 = plt.subplot2grid((2, 2), (1, 0), colspan=1)  # Convolucin
    ax4 = plt.subplot2grid((2, 2), (1, 1), colspan=1)  # Convolucin recortada

    # Graficar la seal 1
    ax1.plot(t, s1, color="r", label="$\sin(2\pi f_1 t) + \sin(2\pi f_2 t)$")
    ax1.set_title("Seal 1")
    ax1.set_xlabel("Tiempo (s)")
    ax1.set_ylabel("Amplitud")
    ax1.legend(loc="lower left")
    ax1.grid(True)

    # Graficar la seal 2
    ax2.plot(t, s2, color="b", label="$\sin(2\pi f_3 t)$")
    ax2.set_title('Seal 2')
    ax2.set_xlabel("Tiempo (s)")
    ax2.set_ylabel("Amplitud")
    ax2.legend()
    ax2.grid(True)

    # Graficar la convolucin
    ax3.plot(t_2, conv, color="g")
    ax3.set_title('Convolucin')
    ax3.set_xlabel("Tiempo (s)")
    ax3.set_ylabel("Amplitud")
    ax3.grid(True)

    # Graficar la convolucin recortada
    ax4.plot(t_2, conv_2, color="black")
    ax4.set_title('Convolucin recortada en el tiempo')
    ax4.set_xlim(0, 1)
    ax4.set_xlabel("Tiempo (s)")
    ax4.set_ylabel("Amplitud")

    # Ajustar diseo y mostrar
    plt.tight_layout()
    plt.show()
```
